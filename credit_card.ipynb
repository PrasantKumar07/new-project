```python
plt.figure(figsize=(6,4))
ax = sns.countplot(x="type", data=df, order=df["type"].value_counts().index)

total = len(df)
for container in ax.containers:
    ax.bar_label(container, labels=[f"{(v/total*100):.1f}%" for v in container.datavalues])

plt.title("Transaction Types (Percentage)")
plt.show()
```

```python
#Correlation matrix
num_cols = df.select_dtypes(include=[np.number]).columns
plt.figure(figsize=(10,8))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Fraud rate by transaction type
if "isFraud" in df.columns and "type" in df.columns:
    fraud_by_type = df.groupby("type")["isFraud"].mean().sort_values(ascending=False)
    print(fraud_by_type)

    plt.figure(figsize=(6,4))
    fraud_by_type.plot(kind="bar")
    plt.title("Fraud Rate by Transaction Type")
    plt.show()
```

```python
print("\n--- Correlation Analysis ---")
# Calculate the correlation matrix for numerical columns.
correlation_matrix = df.corr(numeric_only=True)

# Display the correlation values with the 'isFraud' column, sorted.
print(correlation_matrix['isFraud'].sort_values(ascending=False))
```
